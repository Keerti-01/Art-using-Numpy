# Art using Numpy


```python
import numpy
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
a=numpy.zeros((500,500,3))
```

```python
plt.imshow(a)
```

```python
rect=cv.rectangle(a,(150,150),(300,400),[0,255,0],5)
```

```python
plt.imshow(rect)
```

```python
rect=cv.rectangle(rect,(200,150),(350,400),[0,255,0],5)
```

```python
plt.imshow(rect)
```

```python
new=rect.copy()
```

```python
new=cv.circle(new,(250,100),50,color=[0,255,0],thickness=5)
```

```python
plt.imshow(new)
```

```python
new1=new.copy()
```

```python
new1=cv.circle(new1,(200,450),50,[0,255,0],5)
```

```python
plt.imshow(new1)
```

```python
new1=cv.circle(new1,(300,450),50,[0,255,0],5)
```

```python
plt.imshow(new1)
```

```python
new2=new1.copy()
```

```python
new2=cv.circle(new2,(200,450),25,[0,255,0],5)
```

```python
plt.imshow(new2)
```

```python
new2=cv.circle(new2,(300,450),25,[0,255,0],5)
```

```python
plt.imshow(new2)
```

```python
img=new2.copy()
```

```python
img=cv.line(img,(150,200),(100,250),[0,255,0],5)
```

```python
plt.imshow(img)
```

```python
img=cv.line(img,(100,250),(50,150),[0,255,0],5)
```

```python
plt.imshow(img)
```

```python
img=cv.line(img,(350,200),(450,300),[0,255,0],5)
```

```python
plt.imshow(img)
```
